```python
# In-situ Agro-Ecological Planning Model
# Author: Bismillah Initiative / Research Collaboration
# Purpose: Generate preliminary intervention outlines for semi-arid and tropical regions
# Regions: Palestine, Sudan, Bengal/Myanmar, Sahel, Swahili Coast, Somalia
# -----------------------
# Step 1. Define regional data
# -----------------------

data = {
    "Region": [
        "Palestine", "Sudan", "Bengal/Myanmar", "Sahel", "Swahili Coast", "Somalia"
    ],
    "PrimaryCrops": [
        "Olives, Chickpeas, Barley, Lentils, Fig, Moringa",
        "Sorghum, Millet, Groundnut, Sesame, Baobab",
        "Rice, Jute, Lentil, Banana, Moringa",
        "Millet, Cowpea, Groundnut, Sorghum, Acacia",
        "Cassava, Coconut, Banana, Legumes, Mango",
        "Sorghum, Maize, Cowpea, Cactus Pear, Moringa"
    ],
    "Livestock": [
        "Sheep, Goats, Bees, Poultry",
        "Cattle, Goats, Camels",
        "Fish, Ducks, Goats",
        "Goats, Sheep, Camels",
        "Goats, Cattle, Poultry, Fish",
        "Camels, Goats, Sheep"
    ],
    "WaterStress": [0.8, 0.9, 0.5, 0.85, 0.6, 0.95],  # 0â€“1 scale (1 = very dry)
    "SoilFertility": [0.5, 0.45, 0.7, 0.4, 0.55, 0.35],
    "Accessibility": [0.6, 0.5, 0.65, 0.4, 0.55, 0.45]  # 1 = easy logistics
}

df = pd.DataFrame(data)

# -----------------------
# Step 2. Synthetic Labels (Training Target)
# -----------------------

# Labels = "Intervention Type"
labels = [
    "Tied Ridge + Legume Integration",
    "Tied Ridge + Baobab-Moringa Agroforestry",
    "Floating Rice Beds + Duck Integration",
    "Contour Bunds + Groundnut Intercrop",
    "Coconut-Pigeon Pea Agroforestry + Mulching",
    "Cactus-Moringa Fodder + Sand Filter Water Harvest"
]

df["Label"] = labels

# -----------------------
# Step 3. Encode and Train Model
# -----------------------

X = df[["WaterStress", "SoilFertility", "Accessibility"]]
y = df["Label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(random_state=42, n_estimators=50)
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
report = classification_report(y_test, y_pred, output_dict=True)

# -----------------------
# Step 4. Generate Recommendations
# -----------------------

def make_plan(row):
    if row["WaterStress"] > 0.8:
        water_method = "Tied Ridge / Cactus-based filtration"
    elif row["WaterStress"] > 0.6:
        water_method = "Contour Bunds / In-situ harvesting"
    else:
        water_method = "Rainfed / Pond storage"

    if row["SoilFertility"] < 0.5:
        soil_method = "Legume Rotation + Compost + Fish Waste"
    else:
        soil_method = "Mulch + Crop Residue Integration"

    note = f"Combine {water_method} with {soil_method} and local livestock manure reuse."

    return note

df["RecommendedPlan"] = df.apply(make_plan, axis=1)
df["Predicted_Intervention"] = model.predict(X)

# -----------------------
# Step 5. Output Results
# -----------------------

df.to_csv("region_plans.csv", index=False)

# Save model weaknesses
weakness_text = """
# Model Weaknesses and Recommendations

**1. Data Limitations**
- The dataset is small (6 records), synthetic, and lacks measured field variables (soil pH, salinity, rainfall, cost data).
- Limited representation of seasonality and socio-economic diversity.

**2. Model Constraints**
- RandomForest cannot extrapolate unseen environmental conditions.
- Overfitting possible due to tiny dataset.
- Labels (intervention types) are heuristic, not empirically validated.

**3. Improvement Path**
- Collect 50â€“100 real records per agroecological zone.
- Add dynamic soil-moisture, NDVI, and water-quality indicators.
- Implement participatory validation with local NGOs and farmers.
- Expand to multi-label classification (one region = several possible interventions).
"""

with open("model_weaknesses.md", "w") as f:
    f.write(weakness_text)

# Save report
with open("model_metrics.json", "w") as f:
    json.dump(report, f, indent=4)

print("âœ… Model complete. Outputs:")
print("- region_plans.csv")
print("- model_weaknesses.md")
print("- model_metrics.json")
df
```
